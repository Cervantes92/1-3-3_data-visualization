<h1>Data Visualization Methods</h1>
<p>Here we shall explore methods of data visualization using data from </p>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display as dsp
%matplotlib inline

tb = pd.read_csv('data/TB_Success_Rates.csv')
dsp(tb.head(3))

lit = pd.read_csv('data/Literacy_Rates.csv')
dsp(lit.head(3))

print('Subgroups:\n', lit['Subgroup'].unique())
print('\nCountries Represented:\n', lit['Country or Area'].unique())
print('\nYears represented:\n', lit['Year'].unique())

lit.drop((lit[lit['Country or Area'] == 'fnSeqID'].index) | (lit[lit['Country or Area'] == '1'].index), inplace = True)

sns.lmplot('Year', 'Value', data = lit, fit_reg = False)
plt.title('All data plotted')
plt.ylabel('Literacy Rate (%)')
plt.show()

#dataset = sns.load_dataset('dataset_name')
#Pulls dataset from https://github.com/mwaskom/seaborn-data
```

<h3>Scatterplot of all Measurements</h3>
<p>Here is a scatterplot of all the measruements of literacy for all countries over all years. This scatterplot is neither a clean nor informative way to display the data. Which points represent which countries? Which subgroups? What countries have improved, and what countries have not? We can gleam from this chat a general upward trend in literacy given the density of points in the 2000-2005 80-100% area.</p>

```python
lit_col = lit[lit['Country or Area'] == 'Colombia']

col_f = lit_col[lit_col['Subgroup'] == 'Female 15+ yr']
col_m = lit_col[lit_col['Subgroup'] == 'Male 15+ yr']

dsp(col_f.head())
dsp(col_m.head())

#Plot histogram
plt.title('Histogram of Literacy Measurements in Colombia')
plt.hist(col_f['Value'], color = 'red', alpha = 0.5, label = 'Female', normed = True)
plt.hist(col_m['Value'], color = 'blue', alpha = 0.5, label = 'Male', normed = True)
plt.xlabel('Measurements (%)')
plt.legend(loc = 'upper left')
plt.show()
```

<h3>The Histogram Plot</h3>
<p>This histogram plot of a single country gives more (time independant) information regarding the literacy rates than the earlier scatter plot, but the information layed out is fairly unintuitive. The numbers are also close enough between the two populations that they are hardly distinguishable. A better plot for this particular data set would be two-population literacy rates against time. </p>

```python
plt.figure(figsize = (10,3))

plt.subplot(1,2,1)
plt.plot(col_f['Year'], col_f['Value'],  color = 'red', label = 'Female')
plt.plot(col_m['Year'], col_m['Value'], color = 'blue', label = 'Male')

plt.ylim(0, 100)
plt.title('Annual Literacy Rates')
plt.xlabel('Year')
plt.ylabel('Literacy Rates (%)')
plt.legend(loc = 'lower right')


plt.subplot(1,2,2)
plt.plot(col_f['Year'], col_f['Value'],  color = 'red', label = 'Female')
plt.plot(col_m['Year'], col_m['Value'], color = 'blue', label = 'Male')

plt.title('Annual Literacy Rates')
plt.xlabel('Year')
plt.ylabel('Literacy Rates (%)')
plt.legend(loc = 'lower right')

plt.tight_layout()
plt.show()
```

<h3>The Scatter Plot</h3>
<p>This gives much more useful information than the histogram. We can see how the literacy rate has evolved over time. Unfortunately, there are two issues with this particular graph style.</p>
<ul>
<li>
When we set the y-limits of our graph to be the smallest and largest values in the measured set, the graph implies that inflates the significance of relatively small changes.
</li>
<li>
When we set the y-limits to be the range of all possible measurements (ie 0% to 100%), some of the significance is lost in the visual representation of the data, and it adds to the empty white space.
</li>
</ul>

<p>We will now look at data that is more conducive to histograms and frequency measurements.</p>

```python
#Data from 2000
lit2k = lit[lit['Year'] == 2000]
lit2k_f = lit2k[lit['Subgroup'] == 'Female 15+ yr']
lit2k_m = lit2k[lit['Subgroup'] == 'Male 15+ yr']

plt.hist(lit2k_f['Value'], color = 'red', alpha = 0.5, label = 'Female', normed = True)
plt.hist(lit2k_m['Value'], color = 'blue', alpha = 0.5, label = 'Male', normed = True)

plt.legend(loc = 'upper left')
plt.xlabel('Literacy Rates (%)')
plt.show()
```

<h3>Two Sample Histogram</h3>
<p>Here is the histogram for all measurements taken in 2004 for men and women. Given that we are only looking at two sample groups in one year, this histogram is a reasonable representation of the data on hand. But what happens if we want to look at different Years?</p>

```python
lit2k4 = lit[lit['Year'] == 2004]
lit2k4_f = lit2k[lit['Subgroup'] == 'Female 15+ yr']
lit2k4_m = lit2k[lit['Subgroup'] == 'Male 15+ yr']

plt.hist(lit2k_f['Value'], color = 'red', alpha = 0.5, label = 'Female', normed = True)
plt.hist(lit2k_m['Value'], color = 'blue', alpha = 0.5, label = 'Male', normed = True)
plt.hist(lit2k4_f['Value'], color = 'green', alpha = 0.5, label = 'Female', normed = True)
plt.hist(lit2k4_m['Value'], color = 'yellow', alpha = 0.5, label = 'Male', normed = True)

plt.legend(loc = 'upper left')
plt.xlabel('Literacy Rates (%)')
plt.show()
```

<p>Yech, with all of the overlap, a lot of the data is indistinguishable. We need a plot that visually separates the data by year as well as by sex. We need a box plot!</p>

```python
lit2kor2k4 = lit[(lit['Year'] == 2000) | (lit['Year'] == 2004)]
ax = sns.boxplot(x = 'Year', y = 'Value', hue = 'Subgroup', data = lit2kor2k4)
sns.despine(offset = 10, trim = True)
plt.show()
```

<p>As we can see, this gives us a pretty good understanding of the mean and variance for both years in comparison. Unfortunately, not all countries are represented each year. So what countries are represented from year to year changes. This can have a drastic impact on the visual implications of the dataset. For example, what happens when we apply this graph to all years?</p>

```python
plt.figure(figsize = (30, 10))
ax = sns.boxplot(x = 'Year', y = 'Value', hue = 'Subgroup', data = lit)
sns.despine(offset = 10, trim = True)
plt.show()
```
